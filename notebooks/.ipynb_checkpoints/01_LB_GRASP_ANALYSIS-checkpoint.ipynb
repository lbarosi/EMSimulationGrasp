## Variando FEED

### Gerando Dataset

```python
params_x = [-30, -20, -10, 0, 10, 20, 30]
params_z = [-30, -20, -10, 0, 10, 20, 30]
params_y = [-30, -20, -10, 0, 10, 20, 30]
```

```python
tor_files = []
for param in params_x:
    tor_file = grasp.make_tor(type="translation", value=param)
    tor_files.append(tor_file)
```

```python
# Roda o grasp em background para cada tor file.
# Aguarde o final da execução do GRASP antes de prosseguir.
for tor_file in tor_files:
    job = grasp.run_grasp(tor_file)
    job.pid
```

```python
tor_files = []
for param in params_z:
    tor_file = grasp.make_tor(type="translation", coord="z", value=param)
    tor_files.append(tor_file)
```

```python
# Roda o grasp em background para cada tor file.
# Aguarde o final da execução do GRASP antes de prosseguir.
for tor_file in tor_files:
    job = grasp.run_grasp(tor_file)
    job.pid
```

```python
tor_files = []
for param in params_z:
    tor_file = grasp.make_tor(type="translation", coord="z", value=param)
    tor_files.append(tor_file)
```

```python
# Roda o grasp em background para cada tor file.
# Aguarde o final da execução do GRASP antes de prosseguir.
for tor_file in tor_files:
    job = grasp.run_grasp(tor_file)
    job.pid
```

### Translação

#### Eixo X

```python
# Lê todos os arquivos dual_cut nos diretórios adequados e produz dataframes com resultados
# Lista de dataframes para cada arquivo
# Dataframe com valores de ganho calculados para todos os parâmetros
cut_files = glob.glob("../grasp/STANDARD/job_feed_translation_x*/dual_cut*", recursive=True)
dfs_x = []
gains_x = []
for file in cut_files:
    df = grasp.read_cut(file)
    dfs_x.append(df)
    gains_x.append(grasp.gain_dB(df))
df_gains_x = pd.DataFrame({"Values": params_x, "Gain": gains_x})
```

```python
# Normalizando pelo ganho mínimo e observando a variação
fig, ax = plt.subplots()
ax.plot(df_gains_x.Values, df_gains_x.Gain)
ax.grid()
plt.show();
```

```python
# VIsualização da coleção de padrões de feixe.
ax = grasp.plot_beam_pattern(dfs_x[0], label=params_x[0]);
for ii, df in enumerate(dfs_x):
    grasp.plot_beam_pattern(df, ax=ax, label=params_x[ii]);
#ax.set_xlim([-5, 5])
plt.show();
```

#### Eixo z

```python
cut_files = glob.glob("../grasp/STANDARD/job_feed_translation_z*/dual_cut*", recursive=True)
dfs_z = []
gains_z = []
for file in cut_files:
    df = grasp.read_cut(file)
    dfs_z.append(df)
    gains_z.append(grasp.gain_dB(df))
df_gains_z = pd.DataFrame({"Values": params_z, "Gain": gains_z})
```

```python
fig, ax = plt.subplots()
ax.plot(df_gains_z.Values, df_gains_z.Gain, label="translação eixo z")
ax.plot(df_gains_x.Values, df_gains_x.Gain, label="translação eixo x")
ax.set_xlabel("Deslocamento (cm)")
ax.set_ylabel(r"$\frac{G}{G_0}$ (dB)")
ax.legend()
ax.grid()
plt.show();
```

```python
ax = grasp.plot_beam_pattern(dfs_z[0], label=params_z[0]);
for ii, df in enumerate(dfs_z):
    grasp.plot_beam_pattern(df, ax=ax, label=params_z[ii]);
ax.set_xlim([-5, 5])
plt.show();
```

#### Eixo y

```python
cut_files = glob.glob("../grasp/STANDARD/job_feed_translation_y*/dual_cut*", recursive=True)
dfs_y = []
gains_y = []
for file in cut_files:
    df = grasp.read_cut(file)
    dfs_y.append(df)
    gains_y.append(grasp.gain_dB(df))
df_gains_y = pd.DataFrame({"Values": params_y, "Gain": gains_y})
```

```python
fig, ax = plt.subplots()
ax.plot(df_gains_z.Values, df_gains_z.Gain, label="translação eixo z")
ax.plot(df_gains_y.Values, df_gains_y.Gain, label="translação eixo y")
ax.plot(df_gains_x.Values, df_gains_x.Gain, label="translação eixo x")
ax.set_xlabel("Deslocamento (cm)")
ax.set_ylabel(r"$\frac{G}{G_0}$ (dB)")
ax.legend()
ax.grid()
plt.show();
```
